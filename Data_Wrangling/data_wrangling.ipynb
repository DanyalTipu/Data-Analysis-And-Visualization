### **8. Outliers**

1.   Check if there are any outlier
2.   Handle outliers



```python
normalized_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(normalized_df[:])
```

```python
import pandas as pd
import numpy as np


columns_to_check = normalized_df.columns

def detect_outliers_iqr(data, column):
    Q1 = np.percentile(data[column], 25)
    Q3 = np.percentile(data[column], 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

outliers = pd.DataFrame()
for column in columns_to_check:
    outliers = pd.concat([outliers, detect_outliers_iqr(normalized_df, column)])

print(outliers)
```

```python
normalized_df_clean = normalized_df.drop(index=outliers.index)
print(normalized_df_clean)
```
