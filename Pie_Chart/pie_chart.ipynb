QUESTION 3

```python
import plotly.express as px
fig = px.pie(data, values='Deaths', names='Bldg #', hole=0.5)
fig.show()
```

```python
fig = px.pie(data, values='Deaths', names='Bldg #', color='SpeciesCode')
fig.show()
```

QUESTION 4

```python
import plotly.graph_objects as go
fig = go.Figure(data=[go.Pie(labels=data['Bldg #'], values=data['Deaths'])])
fig.update_traces(textinfo='label+percent', textfont=dict(color='purple'))

fig.show()
```

```python
fig = go.Figure(data=[go.Pie(labels=data['Bldg #'], values=data['Deaths'], hoverinfo='label+percent')])
fig.show()
```

```python
fig = go.Figure(data=[go.Pie(labels=data['Bldg #'], values=data['Deaths'])])
fig.update_traces(hoverinfo='none')
fig.show()
```
