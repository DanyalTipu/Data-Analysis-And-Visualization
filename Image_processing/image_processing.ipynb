```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def rgb_to_gray(img):
        grayImage = np.zeros(img.shape)
        R = (np.array(img[:, :, 0]) *0.3)
        G = (np.array(img[:, :, 1]) *.59)
        B = (np.array(img[:, :, 2]) *.11)

        Avg = (R+G+B)
        grayImage = img.copy()

        for i in range(3):
           grayImage[:,:,i] = Avg

        return grayImage

image = mpimg.imread("/content/Lenna_(test_image).png")
grayImage = rgb_to_gray(image)
plt.imshow(grayImage)
plt.show()
```

```python
grey=grayImage.tolist()
import json
with open("sample.json", "w") as outfile:
    json.dump(grey, outfile)
```

```python
enstien = cv2.imread("/content/einstienn.jpeg")
resizegray_img=cv2.resize(enstien, (512,512))
```

```python
import matplotlib.pyplot as plt
plt.imshow(resizegray_img)
plt.show()
```

```python
gg=np.array(resizegray_img)

def make_histogram(gg,size):
    pixel_value_counts = np.histogram(gg.flatten(), bins=size, range=[0, size-1])[0]
    cumulative_histogram = np.cumsum(pixel_value_counts)
    min_count = np.min(pixel_value_counts)
    max_count = np.max(pixel_value_counts)
    equalized_histogram = ((cumulative_histogram - min_count) * 255) / (max_count - min_count)
    counts_array = np.array(pixel_value_counts)

    return counts_array,equalized_histogram
y,equi=make_histogram(gg,256)
```

```python
plt.hist(y,bins=256)
```
